```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
iris_df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)   # converting it into 2D data
iris_pca_x = pca.fit_transform(iris["data"])
iris_pca_df = pd.DataFrame(data= np.c_[iris_pca_x, iris['target']],
                     columns= ['x', 'y', 'target'])
```

```python
import seaborn as sns
```

```python
pca_graph = sns.lmplot(x='x', y='y', data=iris_pca_df, hue='target', fit_reg=False)
```

```python
graph = sns.lmplot(x='petal length (cm)', y='petal width (cm)', data=iris_df, hue='target', fit_reg=False)
```
